```python
import copy

##浅复制和深复制
print("=====浅复制=========")
a = [1,2,3,[888,999]]
b = a.copy()
print(a,b)
b[3].append(10)
print(a,b)

###深复制
print("=====深复制=========")
a = [1,2,3,[888,999]]
b = copy.deepcopy(a)
print(a,b)
b[3].append(10)
print(a,b)
```

```python
##继承和多态
class Man:
    pass

class Beijing(Man):
    def say():
        print("beijing")

class shanghai(Man):
    def say():
        print("shanghai")
               
```

```python
##赋值、浅复制、深复制
import copy

a = [1,2,3]
b = a
c = copy.copy(a)
d = copy.deepcopy(a)

print("原地址",id(a))
print("赋值地址",id(b))
print("浅复制地址",id(c))
print("深复制地址",id(d))
```
