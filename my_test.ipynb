```python
import copy

##浅复制和深复制
print("=====浅复制=========")
a = [1,2,3,[888,999]]
b = a.copy()
print(a,b)
b[3].append(10)
print(a,b)

###深复制
print("=====深复制=========")
a = [1,2,3,[888,999]]
b = copy.deepcopy(a)
print(a,b)
b[3].append(10)
print(a,b)
```

```python
##赋值、浅复制、深复制
import copy

a = [1,2,3]
b = a
c = copy.copy(a)
d = copy.deepcopy(a)

print("原地址",id(a))
print("赋值地址",id(b))
print("浅复制地址",id(c))
print("深复制地址",id(d))
```

```python
import time 

createTime = '2021-11-20 8:00:04'
now_time = time.strptime(createTime,"%Y-%m-%d %H:%M:%S")
now_minute = now_time.tm_min
now_hour = now_time.tm_hour

if now_minute < 45:
    end_hour = now_hour
else:
    end_hour = now_hour + 1
end_time = createTime[:10]+" "+"{0}:55:00".format(end_hour)
##转时间戳格式
end_struck = time.strptime(end_time,"%Y-%m-%d %H:%M:%S")
end_stamp = int(time.mktime(end_struck))
end_stamp
```

```python
def stamp2date(data):
    """ 时间戳转日期：年月日 """
    struck = time.localtime(data)
    date1 = time.strftime("%Y-%m-%d %H:%M:%S",struck)
    return date1

def Data2Struck(data):
    # 将data转为str格式
    str_data = str(data)
    # 判断是否包含时分秒
    try:
        struck = time.strptime(str_data,"%Y-%m-%d")
    except:
        struck = time.strptime(str_data,"%Y-%m-%d %H:%M:%S")
    return struck

def Data2Timestamp(data):
    """ 字符串日期格式转时间戳格式 """
    struck = Data2Struck(data) 
    stamp = time.mktime(struck)
    return stamp

stamp2date(end_stamp)
```
